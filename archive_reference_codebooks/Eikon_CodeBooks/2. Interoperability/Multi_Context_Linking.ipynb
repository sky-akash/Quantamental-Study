## Multi Context Linking

```python
from refinitiv.workflow import context
```

### Trivial example

```python
# Setting the context
# The first RIC in the list is treated as active_ric
context.set_context(['CHF=', 'EUR=', 'GBP=', '.DJI', '/.SPX', '/EPAM.K'])
```

```python
# Getting the context
context.get_context()
```

### Interactive examples with ipywidgets

```python
from ipywidgets import Output, Button, Text, HBox

# Getting the context
button_get = Button(description='Get')
output_get = Output()

def get_context(_):
    with output_get:
        print(f'rics = {context.get_context().rics}', end=' | ')
        print(f'active_ric = {context.get_context().active_ric}')
        
print(f'Press Get button to get and print the context')
display(button_get, output_get)
button_get.on_click(get_context)
```

```python
# Setting the context
from ast import literal_eval as str_to_list

button_set = Button(description='Set')
text_set = Text()
box_set = HBox([text_set, button_set])
output_set = Output()

def set_context(_):
    context.set_context(str_to_list(text_set.value))
    with output_set:
        print(f'The context was set to {text_set.value}')
        
print(f'Input list of RICs and press Set button to set the context')
display(box_set, output_set)
button_set.on_click(set_context)
```
